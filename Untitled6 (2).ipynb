```python
def lifestyle_score(row):
    # diet
    if row['diet_quality'] == 'Poor':
        diet = 0
    elif row['diet_quality'] == 'Fair':
        diet = 1
    else:  # Good
        diet = 2

    # exercise
    if row['exercise_frequency'] <= 1:
        exercise = 0
    elif 2 <= row['exercise_frequency'] <= 4:
        exercise = 1
    else:  # 5-6
        exercise = 2

    # sleep
    if row['sleep_hours'] < 6:
        sleep = 0
    elif 6 <= row['sleep_hours'] <= 8:
        sleep = 1
    else:
        sleep = 2


    total = diet + exercise + sleep

    if total <= 2:
        return 'Poor'
    elif 3 <= total <= 4:
        return 'Fair'
    else:
        return 'Good'

data['life_style'] = data.apply(lifestyle_score, axis=1)
```

```python

x=data.drop('exam_score',axis=1)
y=data['exam_score']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

```python
coefficients = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': model.coef_
})
print(coefficients)
```

**Model Performance**

```python
coefficients = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': model.coef_
}).sort_values(by='Coefficient', ascending=False)

plt.figure(figsize=(8,6))
plt.barh(coefficients['Feature'], coefficients['Coefficient'])
plt.xlabel("Coefficient Value")
plt.title("Feature Impact on Target")
plt.show()
```
